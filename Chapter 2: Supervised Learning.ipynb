## 1. Konsep Generalization, Overfitting, dan Underfitting
Tujuan utama *Machine Learning* adalah **Generalization**, yaitu kemampuan model memprediksi data baru secara akurat. Jika model terlalu fokus pada *noise* di data latih, akan terjadi **Overfitting**. Jika model terlalu sederhana, terjadi **Underfitting**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline
```

## 2. Analisis Kompleksitas Model: k-Nearest Neighbors
Kita akan menggunakan dataset **Breast Cancer** untuk melihat bagaimana jumlah *neighbors* (k) mempengaruhi performa model. Nilai *k* yang kecil (misal 1) membuat model sangat kompleks (rentan overfitting), sedangkan nilai *k* yang besar membuat model lebih mulus/sederhana (rentan underfitting).

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 11)

# Looping untuk k = 1 hingga 10
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

# Visualisasi Overfitting vs Underfitting
plt.plot(neighbors_settings, training_accuracy, label="Training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="Test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.title("Pengaruh K terhadap Akurasi");
```

Dari plot di atas, performa terbaik pada *Test set* (kemampuan Generalization) didapat ketika `n_neighbors` bernilai sekitar 6.

## 3. Linear Models: Logistic Regression
Meskipun namanya mengandung kata "Regression", **Logistic Regression** adalah algoritma **klasifikasi**. Parameter utamanya adalah `C`. Nilai `C` yang tinggi berarti model akan mencoba menyesuaikan data latih seakurat mungkin (berisiko Overfitting), sedangkan `C` yang rendah menekankan pada koefisien yang mendekati nol (Underfitting).

```python
from sklearn.linear_model import LogisticRegression

# Model default (C=1)
logreg = LogisticRegression(max_iter=10000).fit(X_train, y_train)
print("LogisticRegression Default (C=1)")
print(f"Training set score: {logreg.score(X_train, y_train):.3f}")
print(f"Test set score: {logreg.score(X_test, y_test):.3f}\n")

# Model dengan C tinggi (C=100) -> Model lebih kompleks
logreg100 = LogisticRegression(C=100, max_iter=10000).fit(X_train, y_train)
print("LogisticRegression (C=100) -> Cenderung Overfitting")
print(f"Training set score: {logreg100.score(X_train, y_train):.3f}")
print(f"Test set score: {logreg100.score(X_test, y_test):.3f}\n")

# Model dengan C rendah (C=0.01) -> Model lebih sederhana
logreg001 = LogisticRegression(C=0.01, max_iter=10000).fit(X_train, y_train)
print("LogisticRegression (C=0.01) -> Cenderung Underfitting")
print(f"Training set score: {logreg001.score(X_train, y_train):.3f}")
print(f"Test set score: {logreg001.score(X_test, y_test):.3f}")
```

## 4. Decision Trees (Pohon Keputusan)
Decision Trees membangun model klasifikasi atau regresi dalam bentuk struktur pohon. Algoritma ini sangat rentan terhadap *overfitting* karena dapat membuat pohon yang sangat dalam hingga menghafal setiap sampel data. Oleh karena itu, kita mempraktekkan teknik **Pre-pruning** (pemangkasan) dengan membatasi kedalaman maksimum pohon (`max_depth`).

```python
from sklearn.tree import DecisionTreeClassifier

# Decision Tree Tanpa Batas (Overfitting)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Decision Tree Tanpa Batasan (max_depth=None)")
print(f"Akurasi Training: {tree.score(X_train, y_train):.3f} (100% karena hafal data)")
print(f"Akurasi Test: {tree.score(X_test, y_test):.3f}\n")

# Decision Tree dengan Pre-pruning (max_depth=4)
tree_pruned = DecisionTreeClassifier(max_depth=4, random_state=0)
tree_pruned.fit(X_train, y_train)
print("Decision Tree dengan Batasan (max_depth=4)")
print(f"Akurasi Training: {tree_pruned.score(X_train, y_train):.3f}")
print(f"Akurasi Test: {tree_pruned.score(X_test, y_test):.3f} (Lebih baik karena tidak overfit)")
```
