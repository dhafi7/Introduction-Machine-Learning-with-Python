## 1. Concepts of Generalization, Overfitting, and Underfitting
The primary objective of *Machine Learning* is **Generalization**, which is the model's capability to make accurate predictions on new, unseen data. If a model focuses too heavily on the *noise* within the training dataset, it leads to **Overfitting**. Conversely, if the model structure is too simple to capture the underlying pattern, it results in **Underfitting**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline
```

## 2. Analyzing Model Complexity: k-Nearest Neighbors
We will utilize the **Breast Cancer** dataset to observe how the number of *neighbors* (k) influences model performance. A small value of *k* (e.g., 1) creates a highly complex model decision boundary (making it prone to overfitting), whereas a large value of *k* yields a smoother, simpler model boundary (making it prone to underfitting).

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 11)

# Iterate from k = 1 to 10
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

# Visualizing Overfitting vs. Underfitting
plt.plot(neighbors_settings, training_accuracy, label="Training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="Test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.title("Impact of K on Accuracy");
```

Based on the visualization generated above, the optimal performance on the *Test set* (indicating robust Generalization capability) is achieved when `n_neighbors` is configured around 6.

## 3. Linear Models: Logistic Regression
Despite containing the term "Regression" in its designation, **Logistic Regression** is strictly a **classification** algorithm. Its primary regularization hyperparameter is `C`. A higher value of `C` instructs the model to fit the training data as accurately as possible (increasing the risk of Overfitting), whereas a lower value of `C` places heavier emphasis on forcing the coefficients toward zero (increasing the risk of Underfitting).

```python
from sklearn.linear_model import LogisticRegression

# Default model architecture (C=1)
logreg = LogisticRegression(max_iter=10000).fit(X_train, y_train)
print("LogisticRegression Default (C=1)")
print(f"Training set score: {logreg.score(X_train, y_train):.3f}")
print(f"Test set score: {logreg.score(X_test, y_test):.3f}\n")

# Model with high regularization parameter (C=100) -> Higher model complexity
logreg100 = LogisticRegression(C=100, max_iter=10000).fit(X_train, y_train)
print("LogisticRegression (C=100) -> Prone to Overfitting")
print(f"Training set score: {logreg100.score(X_train, y_train):.3f}")
print(f"Test set score: {logreg100.score(X_test, y_test):.3f}\n")

# Model with low regularization parameter (C=0.01) -> Lower model complexity
logreg001 = LogisticRegression(C=0.01, max_iter=10000).fit(X_train, y_train)
print("LogisticRegression (C=0.01) -> Prone to Underfitting")
print(f"Training set score: {logreg001.score(X_train, y_train):.3f}")
print(f"Test set score: {logreg001.score(X_test, y_test):.3f}")
```

## 4. Decision Trees
Decision Trees construct classification or regression models in the form of a hierarchical tree structure. This algorithm is highly susceptible to *overfitting* because it can generate an overly deep structure that memorizes every individual sample in the training set. To mitigate this behavior, we implement **Pre-pruning** techniques by constraining the maximum depth of the tree via the `max_depth` parameter.

```python
from sklearn.tree import DecisionTreeClassifier

# Unconstrained Decision Tree (Prone to Overfitting)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Decision Tree without constraints (max_depth=None)")
print(f"Training Accuracy: {tree.score(X_train, y_train):.3f} (100% due to data memorization)")
print(f"Test Accuracy: {tree.score(X_test, y_test):.3f}\n")

# Decision Tree with Pre-pruning applied (max_depth=4)
tree_pruned = DecisionTreeClassifier(max_depth=4, random_state=0)
tree_pruned.fit(X_train, y_train)
print("Decision Tree with constraint (max_depth=4)")
print(f"Training Accuracy: {tree_pruned.score(X_train, y_train):.3f}")
print(f"Test Accuracy: {tree_pruned.score(X_test, y_test):.3f} (Improved generalizability due to mitigation of overfit)")
```
