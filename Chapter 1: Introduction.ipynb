## 🟢 Chapter 1: Introduction

Pada bab pertama, buku ini mengenalkan konsep dasar *Machine Learning*, perbedaan antara *Supervised* dan *Unsupervised Learning*. Fokus utama praktik di bab ini adalah membangun model **klasifikasi pertama** menggunakan algoritma **k-Nearest Neighbors (k-NN)** pada dataset Bunga Iris.

### 1.1 Memuat Dataset Iris
Dataset Iris adalah dataset klasik di ML yang berisi ukuran kelopak (petal) dan mahkota (sepal) dari tiga spesies bunga Iris.

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()

print("Kunci dari iris_dataset: \n", iris_dataset.keys())
print("\nNama target (spesies): ", iris_dataset['target_names'])
print("Nama fitur: ", iris_dataset['feature_names'])
```

### 1.2 Memisahkan Data (Train & Test Split) dan Melatih Model (k-NN)
Kita perlu memisahkan data menjadi data latih (*training set*) dan data uji (*test set*).

```python
from sklearn.neighbors import KNeighborsClassifier

# Memisahkan data (75% latih, 25% uji)
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

# Inisialisasi model k-NN dengan k=1
knn = KNeighborsClassifier(n_neighbors=1)

# Melatih model
knn.fit(X_train, y_train)

# Mengevaluasi model
akurasi = knn.score(X_test, y_test)
print(f"Akurasi model k-NN pada test set: {akurasi:.2f} (atau {akurasi*100:.0f}%)")
```

## 🔵 Chapter 2: Supervised Learning

Bab dua masuk lebih dalam ke **Supervised Learning**. Beberapa konsep teori terpenting yang dibahas adalah:
1. **Classification vs Regression:** Klasifikasi memprediksi label kelas (kategorikal), regresi memprediksi nilai kontinu (angka).
2. **Generalization, Overfitting, dan Underfitting:** 
   - *Overfitting:* Model terlalu menghafal data latih, tapi gagal pada data baru.
   - *Underfitting:* Model terlalu sederhana, bahkan tidak bisa menangkap pola pada data latih.

### 2.1 Implementasi Decision Tree pada Dataset Kanker Payudara
Sebagai contoh dari Chapter 2, kita akan menggunakan model **Decision Tree** pada dataset *Breast Cancer* (klasifikasi tumor jinak vs ganas) untuk melihat efek *Overfitting*.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier

cancer = load_breast_cancer()
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)

# Membangun Decision Tree tanpa pembatasan (berpotensi Overfitting)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train_c, y_train_c)

print(f"Akurasi pada training set: {tree.score(X_train_c, y_train_c):.3f}")
print(f"Akurasi pada test set: {tree.score(X_test_c, y_test_c):.3f}")
```

*Perhatikan bahwa akurasi training set adalah 1.000 (100%), yang mengindikasikan bahwa pohon keputusan telah menghafal seluruh label data latih (**Overfitting**), sehingga akurasi pada test set lebih rendah.*

### 2.2 Mencegah Overfitting (Pre-pruning)
Kita dapat mencegah *overfitting* pada *Decision Tree* dengan membatasi kedalaman pohon (*max_depth*).

```python
# Membangun ulang model dengan max_depth = 4
tree_pruned = DecisionTreeClassifier(max_depth=4, random_state=0)
tree_pruned.fit(X_train_c, y_train_c)

print(f"Akurasi training set setelah pre-pruning: {tree_pruned.score(X_train_c, y_train_c):.3f}")
print(f"Akurasi test set setelah pre-pruning: {tree_pruned.score(X_test_c, y_test_c):.3f}")
```

Dengan membatasi kedalaman, akurasi pada *training set* turun, tetapi akurasi pada *test set* meningkat! Ini adalah contoh nyata dari upaya meningkatkan **Generalization**.
