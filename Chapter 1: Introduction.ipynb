## 1. Persiapan Environment dan Library
Hal pertama yang dilakukan adalah mengimpor *library* utama yang akan digunakan untuk manipulasi data dan visualisasi.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Agar plot muncul di dalam notebook
%matplotlib inline
```

## 2. Memuat Dataset (Meet the Data)
Kita akan menggunakan **Iris dataset**, dataset klasik dalam klasifikasi *Machine Learning*. Dataset ini berisi pengukuran bunga Iris (panjang/lebar sepal dan petal) beserta spesiesnya.

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()

print("Keys dari iris_dataset: \n", iris_dataset.keys())
print("\nSpesies yang ingin diprediksi: ", iris_dataset['target_names'])
print("Fitur yang digunakan: ", iris_dataset['feature_names'])
```

## 3. Membagi Data (Training dan Testing)
Kita tidak bisa menggunakan data yang sama untuk melatih dan menguji model. Jika dilakukan, model hanya akan "menghafal" (overfitting). Kita membaginya menggunakan `train_test_split` (75% training, 25% testing).

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

print("Bentuk X_train (Data Latih): ", X_train.shape)
print("Bentuk X_test (Data Uji): ", X_test.shape)
```

## 4. Visualisasi Data (Look at Your Data)
Sebelum membuat model, ada baiknya kita melihat data menggunakan *scatter matrix* untuk memastikan kelas-kelas data bisa dipisahkan dengan baik oleh fitur yang ada.

```python
# Konversi array NumPy ke DataFrame pandas untuk plotting
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# Membuat scatter matrix dari dataframe
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10, 10), marker='o',
                           hist_kwds={'bins': 20}, s=60, alpha=.8, cmap='viridis');
```

## 5. Membangun Model Pertama (k-Nearest Neighbors)
Algoritma k-NN adalah algoritma paling sederhana. Untuk memprediksi data baru, k-NN akan mencari titik data terdekat di data latih (tetangga terdekat) dan memprediksi kelas mayoritas dari tetangga tersebut.

```python
from sklearn.neighbors import KNeighborsClassifier

# Inisialisasi model dengan melihat 1 tetangga terdekat (k=1)
knn = KNeighborsClassifier(n_neighbors=1)

# Melatih model dengan data latih
knn.fit(X_train, y_train)
```

## 6. Membuat Prediksi pada Data Baru
Misalkan kita menemukan bunga Iris di alam dengan panjang sepal 5cm, lebar sepal 2.9cm, panjang petal 1cm, dan lebar petal 0.2cm. Bunga jenis apakah ini?

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape:", X_new.shape)

prediction = knn.predict(X_new)
print("Prediksi kelas: ", prediction)
print("Nama spesies prediksi: ", iris_dataset['target_names'][prediction])
```

## 7. Evaluasi Model
Terakhir, kita uji performa model menggunakan `X_test` dan `y_test` yang sudah disisihkan di awal.

```python
y_pred = knn.predict(X_test)

# Menghitung akurasi
akurasi = knn.score(X_test, y_test)
print(f"Akurasi pada test set: {akurasi:.2f} ({akurasi*100:.0f}%)")
```
